```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
print(stopwords.words('english'))
```

```python
nltk.download('punkt')
```

```python
# Initialize Porter stemmer
stemmer = PorterStemmer()
```

```python
def Steem_text(text):
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    # print (tokens)
    return ' '.join(stemmed_tokens)
```

```python
# Function to remove stopwords
def remove_stopwords(text):

  text = text.replace("'", "")
  tokens = word_tokenize(text)
  filtered_tokens = [word.lower() for word in tokens if word.lower() not in stop_words] #Lower is used to normalize al the words make them in lower case
  print('Tokens are:',tokens,'\n')
  return ' '.join(filtered_tokens)
```

```python
#a function to clean the tweets
def clean(text):
   text = re.sub(r"http\S+", " ", text) # remove urls
   text = re.sub(r"RT ", " ", text) # remove rt
   text = re.sub(r"@[\w]*", " ", text) # remove handles
   text = re.sub(r"[\.\,\#_\|\:\?\?\/\=]", " ", text) # remove special characters
   text = re.sub(r'\t', ' ', text) # remove tabs
   text = re.sub(r'\n', ' ', text) # remove line jump
   text = re.sub(r"\s+", " ", text) # remove extra white space
   text = text.strip()
   return text
```

```python
docs_df['processed_text'] = docs_df['text'].apply(clean)
docs_df
```

```python
docs_df['processed_text'] = docs_df['processed_text'].apply(remove_stopwords)
```

```python
docs_df['processed_text'] = docs_df['processed_text'].apply(Steem_text)
docs_df
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np

#load the ELMo model
elmo = hub.load("https://tfhub.dev/google/elmo/3")

#define the sentences
sentences = ["electronic coordinate transformer", "electronic analogue computer","electronic density states"]

#generate ELMo embeddings for the sentences
embeddings = elmo.signatures["default"](tf.constant(sentences))["elmo"]

#get the ELMo embeddings for the words "apple" and "Apple"
sentence1_embedding = embeddings.numpy()[0][0]  # Embedding for "apple" in the first sentence
sentence2_embedding = embeddings.numpy()[1][0]
sentence3_embedding = embeddings.numpy()[2][0]
  # Embedding for "Apple" in the second sentence

#print the embeddings vectors
print("Embedding vector for 'electronic1':", sentence1_embedding)
print("Embedding vector for 'electronic2':", sentence2_embedding)
print("Embedding vector for 'electronic3':", sentence3_embedding)
```

```python
embeddings.numpy().shape
```

```python
#calculate cosine similarity between the embeddings
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

#compare embeddings
similarity_score = cosine_similarity(sentence1_embedding, sentence2_embedding)
print("Cosine similarity between 'electronic1' and 'electronic2':", similarity_score)

similarity_score = cosine_similarity(sentence1_embedding, sentence3_embedding)
print("Cosine similarity between 'electronic1' and 'electronic3':", similarity_score)

similarity_score = cosine_similarity(sentence2_embedding, sentence3_embedding)
print("Cosine similarity between 'electronic2' and 'electronic3':", similarity_score)
```

# **BERT**

```python
query="computer"
query = preprocess(query)
results=tfidf_retr.search(query)
results
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Load the BERT tokenizer and model
bert_tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
bert_model = BertModel.from_pretrained('bert-base-uncased')

# Loop over the top relevant documents
for doc_id in results['docno'].loc[0:5]:
    # Retrieve the text of the current document
    current_document = docs_df.loc[docs_df['docno'] == doc_id, 'text'].iloc[0]

    # Tokenize the document text
    tokenized_text = bert_tokenizer.tokenize(current_document)

    # Convert tokens to token IDs
    token_ids = bert_tokenizer.convert_tokens_to_ids(tokenized_text)

    # Convert token IDs to tensor
    tokens_tensor = torch.tensor([token_ids])

    # Pass the tokens through the BERT model
    with torch.no_grad():
        outputs = bert_model(tokens_tensor)

    # Obtain the embeddings for the query
    query_embeddings = outputs[0][0].numpy()  # Use the output of the first layer

    # Print the embeddings
    print('BERT embeddings for document', doc_id, ': ', query_embeddings)
```
